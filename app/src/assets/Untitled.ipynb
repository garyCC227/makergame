```python
import json
```

```python
file = "./Bengaluru.geojson"

with open(file) as fp:
  data = json.load(fp)
```

```python
features = data['features']
properties = []
for index, fea in enumerate(features):
  properties.append(fea['properties'])
  if index == 10:
    print(len(properties))
    break
```

```python
file = "./map.geojson"

with open(file) as fp:
  new = json.load(fp)
```

```python
news = new['features']
```

```python
from random import random
random() > 0.5
```

```python
new_features = []
for prop, location in zip(properties, news):
  #add verified and servered 
  prop['verified'] = "true" if random() > 0.5 else "false"
  prop['served'] = "true" if random() > 0.5 else "false"
  location['properties'] = prop
  new_features.append(location)
```

```python
len(new_features)
```

```python
temp ={
  "type":"FeatureCollection",
  "features": new_features
}
```

```python
with open('./map2.geojson', 'w') as fp:
  json.dump(temp, fp, indent=4)
```
